```python
df["Transaction Type"].value_counts()
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"])
df["Month"] = df["Transaction Date"].dt.month

df["Month"].value_counts().sort_index()
```

```python
df["Utilization"] = (df["Credit Card Balance"] / df["Credit Limit"]) * 100

df.groupby("Card Type")["Utilization"].mean()
```

```python
import matplotlib.pyplot as plt

df["Gender"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
plt.hist(df["Account Balance"], bins=20)
plt.title("Balance Distribution")
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x="Gender", y="Account Balance", data=df)

plt.title("Gender vs Account Balance")
plt.xlabel("Gender")
plt.ylabel("Account Balance")

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
df["Balance Category"] = pd.cut(
    df["Account Balance"],
    bins=3,
    labels=["Low", "Medium", "High"]
)

df["Balance Category"].value_counts()
```

```python
loan_counts = df["Loan Type"].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    loan_counts,
    labels=loan_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Loan Type Distribution")

plt.show()
```

```python
top10 = df.nlargest(10, "Account Balance")

plt.figure(figsize=(10,5))

plt.bar(top10["First Name"], top10["Account Balance"])

plt.title("Top 10 Customers by Account Balance")
plt.xlabel("Customer")
plt.ylabel("Account Balance")
plt.xticks(rotation=45)

plt.show()
```

```python
city_balance = df.groupby("City")["Account Balance"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
city_balance.plot(kind="bar")

plt.title("Top 10 Cities by Average Account Balance")
plt.xlabel("City")
plt.ylabel("Average Balance")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df["Credit Limit"], df["Account Balance"], alpha=0.6)

plt.title("Credit Limit vs Account Balance")
plt.xlabel("Credit Limit")
plt.ylabel("Account Balance")

plt.show()
```
